```python
import tensorflow as tf
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
series = np.sin((0.1 * np.arange(400))**2)
plt.plot(series)
plt.show()
```

```python
T = 10
D = 1
X = []
Y = []

for t in range(len(series) - T):
  x = series[t : t+T]
  X.append(x)
  y = series[t+T]
  Y.append(y)

X = np.array(X).reshape(-1, T)
Y = np.array(Y)
N = len(X)
print("X.shape : ", X.shape, "Y.shape : ", Y.shape)
```

# **Simple AutoRegressor**

```python
i = tf.keras.layers.Input(shape=(T,))
x = tf.keras.layers.Dense(1)(i)
model = tf.keras.Model(i,x)
model.compile(
    loss = 'mse',
    optimizer = tf.keras.optimizers.Adam(0.01)
)

r = model.fit(
    X[:-N//2] , Y[:-N//2],
    epochs=80,
    validation_data= (X[-N//2:],Y[-N//2:])
)

plt.plot(r.history['loss'],label='loss')
plt.plot(r.history['val_loss'],label='val_loss')
plt.legend()
plt.show()
```

```python
#wrong forcast

validation_target = Y[-N//2 :]
validation_predictions = []

i = -N//2

while len(validation_predictions) < len(validation_target):
  p = model.predict(X[i].reshape(1 , -1))[0, 0]
  i += 1
  validation_predictions.append(p)

plt.plot(validation_target, label='forecast target')
plt.plot(validation_predictions, label='forecast prediction')
plt.legend()
```

```python
#correct way

validation_target = Y[-N//2:]
validation_predictions = []

last_x = X[-N // 2]

while len(validation_predictions) < len(validation_target):
    p = model.predict(last_x.reshape(1,-1))[0,0]
    validation_predictions.append(p)
    last_x = np.roll(last_x, -1)
    last_x[-1] = p

plt.plot(validation_target, label='forecast target')
plt.plot(validation_predictions, label='forecast prediction')
plt.legend()
```

# **Recurrent network**

```python
X = X.reshape(-1,T,1)

i = tf.keras.layers.Input(shape=(T,D))
x = tf.keras.layers.SimpleRNN(10)(i)
x = tf.keras.layers.Dense(1)(x)

model = tf.keras.Model(i,x)

model.compile(
    loss = 'mse',
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.05)
)

r = model.fit(
    X[:-N//2], Y[:-N//2],
    batch_size=32,
    epochs=200,
    validation_data=(X[-N//2:], Y[-N//2:])
)

plt.plot(r.history['loss'],label='loss')
plt.plot(r.history['val_loss'],label='val_loss')
plt.legend()
plt.show()
```

```python
outputs = model.predict(X)
predictions = outputs[:,0]
plt.plot(Y,label = 'targets')
plt.plot(predictions,label = 'predictions')
plt.legend()
```

```python
forcast = []
input_ = X[-N//2]

while len(forcast) < len(Y[-N//2:]):
  f = model.predict(input_.reshape(1,T,1))[0,0]
  forcast.append(f)

  input_ = np.roll(input_, -1)
  input_[-1] = f

plt.plot(Y[-N//2:], label = 'targets')
plt.plot(forcast, label = 'forcast')
plt.title('Simple RNN forcast')
plt.legend()
plt.show()
```

# **LSTM Network**

```python
X = X.reshape(-1,T,1)

i = tf.keras.layers.Input(shape=(T,D))
x = tf.keras.layers.LSTM(10)(i)
x = tf.keras.layers.Dense(1)(x)

model = tf.keras.Model(i,x)

model.compile(
    loss = 'mse',
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.05)
)

r = model.fit(
    X[:-N//2], Y[:-N//2],
    batch_size=32,
    epochs=200,
    validation_data=(X[-N//2:], Y[-N//2:])
)

plt.plot(r.history['loss'],label='loss')
plt.plot(r.history['val_loss'],label='val_loss')
plt.legend()
plt.show()
```

```python
outputs = model.predict(X)
predictions = outputs[:,0]
plt.plot(Y,label = 'targets')
plt.plot(predictions,label = 'predictions')
plt.legend()
```

```python
forcast = []
input_ = X[-N//2]

while len(forcast) < len(Y[-N//2:]):
  f = model.predict(input_.reshape(1,T,1))[0,0]
  forcast.append(f)

  input_ = np.roll(input_, -1)
  input_[-1] = f

plt.plot(Y[-N//2:], label = 'targets')
plt.plot(forcast, label = 'forcast')
plt.title('LSTM forcast')
plt.legend()
plt.show()
```
