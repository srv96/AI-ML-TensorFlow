```python
#minimize the loss function
#L(w) = w ** 2

w = tf.Variable(5.)

def get_loss(w):
  return w ** 2

def get_grad(w):
  with tf.GradientTape() as tape:
    L = get_loss(w)
  g = tape.gradient(L, w)
  return g

optimizer = tf.keras.optimizers.SGD(learning_rate=0.1)
losses = []

epochs = 50
for i in range(epochs):
  g = get_grad(w)
  optimizer.apply_gradients(zip([g] , [w]))
  losses.append(get_loss(w))
```

```python
import matplotlib.pyplot as plt
plt.plot(losses)
print(f'final loss {losses[-1]}')
```

```python
#manual gradient descent with know function
w = tf.Variable(5.)
losses2 = []

epochs = 50
for i in range(epochs):
  w = w - 0.1 * 2 * w
  losses2.append(w ** 2)

plt.plot(losses)
print(f'final loss {losses[-1]}')
```

```python
plt.plot(losses, label = 'los_tf')
plt.plot(losses2 , label = 'los_manual')
plt.legend()
```
